# 확률적 경사 하강법
## KEYWORD : 손실 함수, 에포크
> 경사 하강법의 알고리즘을 이해하고, 대량의 데이터에서 분류 모델을 훈련하는 방법 배우기

---

## 점진적인 학습
> 앞서 훈련한 모델을 버리지 않고 새로운 데이터에 대해서만 조금씩 더 훈련
```
대표적인 알고리즘 : "확률적 경사 하강법"
```

## 확률적 경사 하강법
### 경사 하강법
- 가장 가파른 경사를 따라 원하는 지점에 도달하는 것이 목표
- 단, 천천히 조금씩
- 이 과정이 바로 경사 하강법 모델을 훈련하는 것

### '확률적' 경사 하강법
- 훈련 세트에서 랜덤하게 **하나의 샘플**을 선택하여 가파른 경사를 조금 내려감
- 또 훈련 세트에서 또 다른 샘플을 하나 선택하여 경사를 조금 내려감
- ... 이런 식으로 전체 샘플을 모두 사용할 때까지 계속

### 에포크 (epoch)
- 확률적 경사 하강법에서 훈련 세트를 한 번 모두 사용하는 과정
- 모든 샘플을 다 사용한 뒤에도 원하는 지점에 도달하지 못하면, 훈련 세트에 모든 샘플을 다시 채워 넣고 다시 수행
- 일반적으로 경사 하강법은 수십, 수백 번 이상 에포크를 수행

### 미니배치 경사 하강법
- 여러 개의 샘플을 사용해 경사 하강법을 수행하는 방식

### 배치 경사 하강법
- 한 번 경사로를 따라 이동하기 위해 전체 샘플을 사용
- 가장 안전할 수 있지만, 자원을 많이 사용함

---
## 여기까지 정리
훈련 세트에서 샘플을

  $┗$ 1개씩 꺼내기 : 확률적 경사 하강법<br>
  $┗$ 여러 개씩 꺼내기 : 미니배치 경사 하강법<br>
  $┗$ 몽땅 꺼내기 : 배치 경사 하강법<br>
  $┗$ 한 번 모두 사용하면 : 에포크


> 기억할 것 : 신경망 모델이 확률적 경사 하강법이나 미니배치 경사 하강법을 사용한다는 점

## 손실 함수
> 어떤 문제에서 머신러닝 알고리즘이 얼마나 엉터리인지를측정하는 기준, 최솟값을 알 수 없음

- 손실 함수의 최종 목적은 ** 잘 맞히면 0에 가깝게, 못 맞히면 엄청나게 큰 양수**를 만드는 것

[!] 손실함수와 비용함수
- 손실 함수 : 샘플 하나에 대한 손실을 정의
- 비용 함수 : 훈련 세트에 있는 모든 샘플에 대한 손실 함수의 함
- 보통은 구분하지 않고 섞어서 사용함
